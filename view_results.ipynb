# Contexto Solver — Results Viewer

```python
import json
import glob
import pandas as pd
from IPython.display import display, HTML

# List available result files
files = sorted(glob.glob("results/*.jsonl"))
for i, f in enumerate(files):
    print(f"{i}: {f}")
```

```python
# Pick a file by index (change this to view a different game)
FILE_INDEX = 0

with open(files[FILE_INDEX]) as f:
    data = json.loads(f.readline())

print(f"Game #{data['game_id']}")
print(f"Timestamp: {data['timestamp']}")
print(f"Duration: {data['duration_seconds']:.1f}s")
print(f"Successful: {data['successful']}")
print(f"Best word: {data['best_word']} (score: {data['best_score']:.4f})")
print(f"Total guesses: {len(data['trajectory'])}")
```

```python
# Trajectory table
df = pd.DataFrame(data["trajectory"])

import matplotlib.cm as cm
import matplotlib.colors as mcolors

score_cmap = cm.get_cmap("viridis")

def color_score(val):
    if val <= 0:
        return "background-color: #555; color: #aaa"
    r, g, b, _ = score_cmap(val)
    luminance = 0.299 * r + 0.587 * g + 0.114 * b
    text_color = "white" if luminance < 0.5 else "black"
    return f"background-color: {mcolors.to_hex((r, g, b))}; color: {text_color}"

def color_mode(val):
    colors = {"seed": "#264653", "main": "#2a9d8f", "jump": "#e76f51"}
    bg = colors.get(val, "#333")
    return f"background-color: {bg}; color: white"

styled = (
    df.style
    .map(color_score, subset=["score"])
    .map(color_mode, subset=["mode"])
    .format({"score": "{:.4f}"})
)
display(styled)
```
